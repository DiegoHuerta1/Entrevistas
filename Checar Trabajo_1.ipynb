### Ejemplo: Crear todos los objetos de entrevistas

Guardar en una lista

```python
entrevistas = []

# checar todos
for file in os.listdir('.\\' + carpet):
    ruta_archivo = os.path.join('.\\' + carpet, file)
    if os.path.isfile(ruta_archivo):
        entrevistas.append(Entrevista(carpet, file, debug = 0))
        
print(f'Se tienen {len(entrevistas)} entrevistas')
```

```python
# imprimir resumen por todas las entrevistas
for e in entrevistas:
    e.print_resumen()
```

```python
# Ejemplo, obtener todas las preguntas
preguntas_todo = []

for e in entrevistas:
    preguntas_todo.extend(e.get_dialogos_entrevistador())
    
print(len(preguntas_todo))
preguntas_todo
```
